```python
from jax import config
config.update("jax_enable_x64", True)

import numpy as np
import matplotlib.pyplot as plt
import croissant as cro
import croissant.jax as crojax
import healpy as hp
import jax
import jax.numpy as jnp
from astropy.time import Time

%matplotlib widget
```

```python
npz = np.load("/home/christian/Documents/research/eigsep/eigsep_vivaldi.npz")
beam = npz["bm"]
freqs = npz["freqs"] / 1e6
nside = npz["nside"]
lmax = 3 * nside - 1 
print(nside, lmax)
```

```python
TIME = Time("2025-02-27 10:24:01")
lat, lon = (39.2480857349511, -113.40368874682412)
LOC = (lon, lat)

sky = cro.Sky.gsm(freqs, lmax=lmax)
sky.switch_coords("C")

beam_alm = cro.healpix.map2alm(beam, lmax=lmax)
beam = cro.Beam(beam_alm, lmax=lmax, frequencies=freqs, coord="T")
beam.horizon_cut(nside=nside)
beam.switch_coords("C", loc=LOC, time=TIME)
```

```python
plot_ix = 50
plot_freq = freqs[plot_ix]
hp.projview(beam.hp_map(nside, frequencies=[plot_freq])[0], title=f"Beam at {plot_freq:.0f} MHz")
plt.show()

hp.projview(sky.hp_map(nside, frequencies=[plot_freq])[0], title=f"GSM at {plot_freq:.0f} MHz")
plt.show()
```

```python
sky.alm.shape
```

```python
usky, ssky, vtsky = np.linalg.svd(sky.alm @ sky.alm.T)
ubeam, sbeam, vtbeam = np.linalg.svd(beam.alm @ beam.alm.T)
plt.figure()
plt.plot(ssky/ssky[0], marker="o", ls="none", label="Sky")
plt.plot(sbeam/sbeam[0], marker="o", ls="none", label="Beam")
plt.yscale("log")
plt.legend()
#plt.xlim(0, 25)
#plt.ylim(1e-8, 1.2)
plt.show()
```

```python
thresh = 1e-8
NSKY = np.count_nonzero(ssky/ssky[0] >= thresh)
NBEAM = np.count_nonzero(sbeam/sbeam[0] >= thresh)
print(NSKY, NBEAM)
sky_modes = vtsky[:NSKY]
beam_modes = vtbeam[:NBEAM]

fig, axs = plt.subplots(nrows=2, constrained_layout=True, sharex=True, sharey=True)
axs[0].plot(freqs, sky_modes.T)
axs[1].plot(freqs, beam_modes.T)
plt.show()
```

```python
from astropy import units as u
```

```python
t_end = TIME + cro.constants.sidereal_day_earth * u.s
times = cro.utils.time_array(t_start=TIME, t_end=t_end, N_times=100)
sim = cro.Simulator(beam, sky, lmax=lmax, frequencies=freqs, world="earth", times=times)
```

```python
sim.run(dpss=False)
sim.plot()
```

```python
uw, sw, vtw = np.linalg.svd(sim.waterfall.T @ sim.waterfall)
vis_modes = np.empty((len(sky_modes) * len(beam_modes), freqs.size))
i = 0
for sm in sky_modes:
    for bm in beam_modes:
        vis_modes[i] = sm * bm
        i += 1

u, s, vt = np.linalg.svd(vis_modes.T @ vis_modes)
plt.figure()
plt.plot(s/s[0], marker="o", ls="none", label="outer product")
plt.plot(sw/sw[0], marker="o", ls="none", label="waterfall")
plt.yscale("log")
#plt.xlim(0, 25)
#plt.ylim(1e-8, 1.2)
plt.legend()
plt.show()


plt.figure()
plt.plot(freqs, vis_modes.T)
plt.show()
```

```python
# m = 0 modes (the ones that survive a full day avg)
plt.figure()
plt.plot(np.arange(lmax), np.abs(sky.alm[0, :lmax])/np.abs(sky[0, 0, 0]))
plt.plot(np.arange(lmax), np.abs(beam.alm[0, :lmax])/np.abs(beam[0, 0, 0]))
plt.yscale("log")
plt.show()
```

```python
NW = np.count_nonzero(sw/sw[0] >= thresh)
print(NSKY, NBEAM, NW)
wf_modes = vtw[:NW]

plt.figure()
plt.plot(freqs, wf_modes.T)
plt.show()
```

```python
mean = sim.waterfall.mean(axis=0)

plt.figure()
plt.plot(freqs, mean)
plt.show()

plt.figure()
plt.plot(np.abs(wf_modes @ mean), marker="o", ls="none")
plt.yscale("log")
plt.show()
```

This just shows that the first mode is basically the mean spectrum. What if we construct the modes without the mean?

```python
wf_nomean = sim.waterfall - mean[None, :]
uw2, sw2, vtw2 = np.linalg.svd(wf_nomean.T @ wf_nomean)

NW2 = np.count_nonzero(sw2/sw2[0] >= thresh)
print(NSKY, NBEAM, NW, NW2)
wf_modes_nomean = vtw2[:NW2]

plt.figure()
plt.plot(freqs, wf_modes_nomean.T)
plt.show()
```
