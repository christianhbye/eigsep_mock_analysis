Trying to reproduce CROISSANT results with a big matrix.

```python
def extend_alm(arr):
    """
    Add negative m-modes.
    Healpy doesn't include the negative m-modes, since they can be derived from the positive ones, this adds them explictly.

    arr can have multiple dimensions as long as the alm-dimension is the last one.
    
    """
    lmax = hp.Alm.getlmax(arr.shape[-1])
    emms = hp.Alm.getlm(lmax)[1]
    m0_modes = arr[..., :lmax+1]
    mpos_modes = arr[..., lmax+1:]
    mneg_modes = (-1)**emms[lmax+1:] * mpos_modes
    arr2 = np.concatenate((m0_modes, mpos_modes, mneg_modes), axis=-1)
    return arr2


def reduce_alm(arr, lmax):
    """
    Invert the function above.
    """
    s = hp.Alm.getsize(lmax)
    return arr[..., :s]
```

```python
phases = sim.sky.rot_alm_z(times=sim.dt, world="earth")
Amat = phases[None, :, :].conj() * beam.alm[:, None, :]  # freq, time, alm

xvec = extend_alm(sky.alm.conj())
Amat = extend_alm(Amat)

y = np.empty((freqs.size, times.size))
for i in range(len(freqs)):
    y[i] = Amat[i] @ xvec[i]

print(np.allclose(y.T/sim.beam.total_power, sim.waterfall))
```

## Mapmaking

Three methods: Per-frequency linear estimator, per-eigenmode linear estimator, JAX.

### Per-frequency (Tegmark 1997 style)

https://ui.adsabs.harvard.edu/abs/1997ApJ...480L..87T/abstract

```python
def m2(Amat, eps=1e-7):
    W = np.empty_like(Amat)
    W = np.swapaxes(W, 1, 2)
    for i, A in enumerate(Amat):
        W[i] = np.linalg.inv(A.T @ A + eps*np.eye(A.shape[-1])) @ A.T
    return W

def wiener(Amat, S, eps=1e-7):
    W = np.empty((Amat.shape[-1], Amat.shape[-2]))
    if Amat.shape[-1] > Amat.shape[-2]:  # use method 4
        for i, A in enumerate(Amat):
            W[i] = S[i] @ A.T @ np.linalg.inv(A @ S[i] @ A.T + eps*np.eye(A.shape[0]))
    else:
        for i, A in enumerate(Amat):
            Sinv = np.diag(1 / np.diag(S[i]))  # assumes S is DIAGonal
            Ninv = 1/eps * np.eye(A.shape[0])
            W[i] = np.linalg.inv(Sinv + A.T @ Ninv @ A) @ A.T @ Ninv
    return W
    
nf = 2

xhat = np.empty_like(xvec)[:nf]
W = m2(Amat[:nf], eps=1e-3)
for i in range(nf):
    xhat[i] = W[i] @ y[i]

sky_hat = cro.healpix.alm2map(reduce_alm(xhat, lmax).conj(), nside, lmax=lmax)
```

### Using GSM eigenmodes

```python
u, s, vt = np.linalg.svd(xvec @ xvec.T.conj())

plt.figure()
plt.plot(s/s[0], marker="o", ls="none", c="k")
plt.yscale("log")
plt.show()
```

```python
Nmodes = 7
sky_modes = vt[:Nmodes].real
plt.figure()
plt.plot(freqs, sky_modes.T)
plt.show()
```

```python
# use sparse linear operators to compute the sky estimator

from scipy.sparse.linalg import LinearOperator, lsqr
Nf, Nt = y.shape
N_alm = Amat.shape[-1]  # extended alm with negative ms
design_shape = (Nf * Nt, Nmodes * N_alm)

r = extend_alm(phases.conj())
b = extend_alm(beam.alm)

def matvec(v):
    """
    Compute the matrix product A @ v for a vector v representing sky-coefficients in the space spanned by GSM-modes and spherical harmonics.

    A is the design matrix (of beam alm and rotation phases) that sends sky to waterfall.

    We assume v is flattened.
    
    """
    vm = v.reshape(N_alm, Nmodes)
    result = np.empty((Nf, Nt))
    for f in range(Nf):
        w = vm @ sky_modes[:, f]  # put sky back in freq space
        result[f] = Amat[f] @ w
    return result.ravel()  # flatten to a vector of length Nf*Nt

def rmatvec(u):
    """
    Adjoint matrix multiply, that is A^H @ u. "From waterfall to sky."

    """
    um = u.reshape(Nf, Nt)
    result = np.zeros((N_alm, Nmodes), dtype=u.dtype)
    for f in range(Nf):
        f1 = b[f].conj() * (um[f] @ r.conj())
        f2 = sky_modes[:, f]
        result += f1[:, None] * f2[None, :]
    return result.ravel()  # Flatten to shape (M*Nmodes,)

    

A_op = LinearOperator(design_shape, matvec=matvec, rmatvec=rmatvec, dtype=np.complex128)
```

```python
sol = lsqr(A_op, y.ravel().astype(complex))
c_est = sol[0]  # estimated solution, shape: (M*Nmodes,)

# Reshape c_est into the coefficient matrix C of shape (M, Nmodes)
C_est = c_est.reshape(N_alm, Nmodes)
```

```python
C_est_hp = reduce_alm(C_est.T.copy(), lmax)
C_est_pix = cro.healpix.alm2map(C_est_hp.conj(), nside, lmax=lmax)
sky_hat_PCA = sky_modes.T @ C_est_pix
```
